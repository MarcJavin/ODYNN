## Neuron simulation

```python
import opthh.neursimul as sim
import scipy as sp
```

```python
t = sp.arange(0., 1200., 0.1)
i = 40. * ((t > 400) & (t < 800))
sim.simul(dt=0.1, i_inj=i, show=True, save=False)
```
