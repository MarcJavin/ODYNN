## Neuron simulation

```python
from opthh.neuronsimul import NeuronSimul
import scipy as sp
```

```python
t = sp.arange(0., 1200., 0.1)
i = 40. * ((t > 400) & (t < 800))
simul = NeuronSimul(t=t, i_inj=i)
simul.simul(show=True, save=False)
```
